```python
from ultralytics import YOLO
import cv2
```

```python
# Load your trained model (use best.pt or last.pt from your training folder)
model = YOLO('C:/continer/work/code/ai Summmer/machine learing/detection Project/drone_detection_run2/train_fast_320/weights/best.pt')
```

```python
# Open webcam
cap = cv2.VideoCapture(0)  # 0 = default webcam

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Resize frame for better visibility (optional)
    frame = cv2.resize(frame, (640, 480))

    # Run YOLOv11 inference with lower confidence threshold
    results = model.predict(source=frame, conf=0.1, verbose=False)

    # Draw bounding boxes and labels
    annotated_frame = results[0].plot()

    # Show the annotated frame
    cv2.imshow("YOLOv11 Webcam Detection", annotated_frame)

    # Exit when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release webcam and close window
cap.release()
cv2.destroyAllWindows()
```
